```python
# libraries
import xarray as xr
import matplotlib.pyplot as plt
import regex as re
```

```python
# load ERA5
store1 = 'gs://gcp-public-data-arco-era5/ar/full_37-1h-0p25deg-chunk-1.zarr-v3'
ds1 = xr.open_dataset(store1, engine='zarr', chunks={})
```

```python
# data exploration
mylist = list(ds1.keys())
r = re.compile(".*precip.*")
newlist = list(filter(r.match, mylist)) # Read Note below
print(newlist)
```

```python
# weather variables
"""
'convective_available_potential_energy'
'convective_inhibition'
'specific_humidity'
'total_precipitation'
'temperature'
'surface_pressure'
"""
```

```python
y_temp = ds1.mean(dim="level")[['temperature']]
```

```python
fig,axes = plt.subplots(figsize=(18,4.5),ncols=3)
yr0, yr1, yr2 = '1900-01-01T00:00:00.000000000', '1950-01-01T00:00:00.000000000', '2000-01-01T00:00:00.000000000'

y_temp.sel(time=yr0).temperature.plot(ax=axes.flat[0],cmap='RdBu_r')
y_temp.sel(time=yr1).temperature.plot(ax=axes.flat[1],cmap='RdBu_r')
y_temp.sel(time=yr2).temperature.plot(ax=axes.flat[2],cmap='RdBu_r')

fig.suptitle('historical temperature data',fontweight='bold')
plt.tight_layout()
```

```python
# slicing
d0, d1 = '2024-01-01T00:00:00', '2025-01-01T00:00:00'
lat_min, lat_max = 55, 20
lon_min, lon_max = 230, 300
y_temp.sel(time=slice(d0,d1),latitude=slice(lat_min,lat_max),longitude=slice(lon_min,lon_max)).temperature
```
